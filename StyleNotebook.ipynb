```python
#importando bibliotecas para utilização
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
#setando n° máximo de linhas para visualização no jupyter
pd.set_option('display.max_rows',200)
```

```python
#lendo dataset do covid em excel
dados_covid = pd.read_excel('C:/Users/RenanAugusto/Documents/Datasets/COVIDBR_12jun2020.xlsx')
```

```python
#dimensão do dataset
dados_covid.shape
```

```python
#visualizando as últimas 5 linhas do dataset
dados_covid.tail(5)
```

```python
#selecionando o município de São Paulo
df_sp = dados_covid.loc[dados_covid.municipio == 'São Paulo']
```

```python
#selecioando as colunas que quero
col_selecionadas = ['data','estado','municipio','casosNovos','obitosNovos','casosAcumulado','obitosAcumulado']
```

```python
#dataframe finalizado
df_sp = df_sp[col_selecionadas]
```

```python
#eliminando a primeira linha pois só tinha zeros
df_sp = df_sp.drop([264976,264977], axis=0)
```

```python
#verificando dataset
df_sp
```

```python
#esconde o index
df_sp.style.hide_index()
```

```python
#atribuindo cor na célula de maior valor para cada coluna numérica
df_sp.style.highlight_max()
```

```python
#pintado a célula MÍNIMA de algumas colunas de vermelho
(
df_sp.style.hide_index()
.highlight_max()
.highlight_min(subset=['casosNovos','obitosNovos'], color='red')
)
```

```python
#adicionando fundo gradiente
(
df_sp.style.hide_index()
.background_gradient('Blues')
)
```

## Podemos encontrar mais mapas de gradientes de cores em:
### https://matplotlib.org/3.1.0/tutorials/colors/colormaps.html

```python
#setando caracteristicas específicas
(
df_sp.style.set_properties(**{'background-color':'black',
                             'color':'lawngreen',
                             'border-color':'white'})
.hide_index()
)
```

```python
#setando configurações da tabela
(
df_sp.style.hide_index()
.set_table_styles([{'selector':'tr:hover',    #colocando barra de sinalização
                   'props':[('background-color','black'),('color','white')]}])
)
```

```python
#criando função para pintar a letra de vermelho caso negativo, ou azul, caso positivo
def color_negative_red(val):
    color = 'red' if val < 0 else 'blue'
    return 'color: %s' % color

df_sp[['casosNovos','obitosNovos','casosAcumulado','obitosAcumulado']].style.applymap(color_negative_red)
```
